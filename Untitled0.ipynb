```python

data = pd.read_csv("/content/Sample Data - Interview Case Study.xlsx - Sample Data (1).csv")

# Converting 'txn_mth' to datetime format and extract year and month
data['txn_mth'] = pd.to_datetime(data['txn_mth'])
data['year'] = data['txn_mth'].dt.year
data['month'] = data['txn_mth'].dt.month

# Converting approval_rate to numeric
data['approval_rate'] = pd.to_numeric(data['approval_rate'], errors='coerce')
data['new_merchant_acct_country_code'].fillna("Unknown",inplace=True)

# Grouping by year, month, and country to calculate the average approval rate
approval_rate_summary = data.groupby(['year', 'month', 'new_merchant_acct_country_code']).approval_rate.mean().reset_index()

# Renaming columns for clarity
approval_rate_summary.columns = ['Year', 'Month', 'Country', 'Average_Approval_Rate']

# Displaying the result
print(approval_rate_summary)
```

   - This lower rate could be due to stricter authorization processes or limits on debit transactions, which might lead to higher declines due to insufficient funds, spending caps, or risk controls.

### Summary Observations:
- **Credit cards are the most favored in terms of approval rates**, likely due to their higher reliability and lower perceived risk.
- **Debit and prepaid cards show lower approval rates**, possibly due to their specific usage constraints and additional checks or limitations that issuers or merchants may place on these types of cards.

## Instead of using Histogram  and Kde (smootens the graph) to find the skewness of the data  we used formula to find the exact skewness  as you can see here there is right skewness in the both Attempts and the Approved attempts data's

```python
# Calculate the first (Q1) and third (Q3) quartiles for each column
Q1_attempts = monthly_trends['attempts'].quantile(0.25)
Q3_attempts = monthly_trends['attempts'].quantile(0.75)
IQR_attempts = Q3_attempts - Q1_attempts

Q1_approved_attempts = monthly_trends['approved_attempts'].quantile(0.25)
Q3_approved_attempts = monthly_trends['approved_attempts'].quantile(0.75)
IQR_approved_attempts = Q3_approved_attempts - Q1_approved_attempts

# Calculate lower and upper bounds for each column
lower_bound_attempts = Q1_attempts - 1.5 * IQR_attempts
upper_bound_attempts = Q3_attempts + 1.5 * IQR_attempts

lower_bound_approved_attempts = Q1_approved_attempts - 1.5 * IQR_approved_attempts
upper_bound_approved_attempts = Q3_approved_attempts + 1.5 * IQR_approved_attempts

# Filter out the anomalies using the IQR method
anomalies_iqr = monthly_trends[
    (monthly_trends['attempts'] < lower_bound_attempts) |
    (monthly_trends['attempts'] > upper_bound_attempts) |
    (monthly_trends['approved_attempts'] < lower_bound_approved_attempts) |
    (monthly_trends['approved_attempts'] > upper_bound_approved_attempts)
]

# Print the anomalies
print(anomalies_iqr)
```

# USED BOX_COX TRANSFORMATION

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Apply Box-Cox transformation to 'attempts' and 'approved_attempts'
# Note: Box-Cox only works for strictly positive values
monthly_trends['attempts_boxcox'], attempts_lambda = stats.boxcox(monthly_trends['attempts'] + 1)  # Add 1 if there are zero values
monthly_trends['approved_attempts_boxcox'], approved_attempts_lambda = stats.boxcox(monthly_trends['approved_attempts'] + 1)  # Add 1 if there are zero values

# Plot the original vs transformed data for comparison
plt.figure(figsize=(12, 6))

# Plot original data
plt.subplot(1, 2, 1)
sns.boxplot(data=monthly_trends[['attempts', 'approved_attempts']])
plt.title('Original Data (Attempts and Approved Attempts)')

# Plot transformed data
plt.subplot(1, 2, 2)
sns.boxplot(data=monthly_trends[['attempts_boxcox', 'approved_attempts_boxcox']])
plt.title('Box-Cox Transformed Data')

plt.tight_layout()
plt.show()

# Print the lambda values (Box-Cox transformation parameters)
print(f"Box-Cox transformation lambda for attempts: {attempts_lambda}")
print(f"Box-Cox transformation lambda for approved_attempts: {approved_attempts_lambda}")
```
